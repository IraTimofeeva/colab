```python
import numpy as np
```

```python
#10 нулей
a=np.zeros(10)
print(a)
```

```python
#10 единиц
a=np.ones(10)
print(a)
```

```python
#10 пятерок
a=np.ones(10)*5
print(a)
```

```python
#Создать массив из целых чисел от 10 до 50
a=np.arange(10,51)
print(a)
```

```python
#Создать массив из четных чисел от 10 до 50
a=np.arange(10,51,2)
print(a)
```

```python
#Создать матрицу 3х3 с числами от 0 до 8
a=np.arange(0,9)
b=a.reshape(3,3)
print(b)
```

```python
#Создать единичную матрицу 3x3
a=np.eye(3)
print(a)
```

```python
#Используйте NumPy для генерации случайного числа от 0 до 1
a=np.random.rand()
print(a)
```

```python
#Используйте NumPy для генерации массива из 25 случайных чисел,взятых из стандартного нормального распределения
a=np.random.rand(25)
print(a)
```

```python
a=np.arange(0.01,1.01,0.01)
b=a.reshape(10,10)
print(b)
```
